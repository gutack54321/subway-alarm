```python
sleep_list=[]
train_code_list=[]
return_time = 1
while return_time <= 30:
    url_2="http://swopenAPI.seoul.go.kr/api/subway/"+api_code_1+"/xml/realtimeStationArrival/0/500/"+sw_end_s
    res_2 = rq.get(url_2)
    soup2 = BeautifulSoup(res_2.text, 'html')


#     btrain_no = soup2.find_all('btrainno') #도착api에서 열차번호
#     btrain_msg = soup2.find_all('arvlmsg2')
    btrain_arvlcd = soup2.find_all('arvlcd')
#     btrain_updnline = soup2.find_all('updnline')

    return_time+=1 


    for i in btrain_arvlcd:  
        arvlcd_txt = i.text 
        arvlcd_txt=int(arvlcd_txt)
        arvlmsg2_txt = i.parent.arvlmsg2.text 
        arvlmsg3_txt = i.parent.arvlmsg3.text 
        updnline_txt = i.parent.updnline.text
        btrainno_txt = i.parent.btrainno.text

        if arvlcd_txt== 3 and updnline_txt== updown_condition and btrainno_txt== train_code: 
            sleep_list.append(arvlmsg3_txt)
            train_code_list.append(btrainno_txt)
            print(f'{arvlmsg2_txt}입니다. 준비합시다')
            break
        elif arvlcd_txt== 4 and updnline_txt== updown_condition and btrainno_txt== train_code: 
            sleep_list.append(arvlmsg3_txt)
            train_code_list.append(btrainno_txt)
            print(f'{arvlmsg2_txt}입니다. 준비합시다')
            break
        elif arvlcd_txt== 5 and updnline_txt== updown_condition and btrainno_txt== train_code: 
            sleep_list.append(arvlmsg3_txt)
            train_code_list.append(btrainno_txt)
            print(f'{arvlmsg2_txt}입니다. 준비합시다')
            break
            
            
            
            
        elif arvlcd_txt== 0 and updnline_txt== updown_condition and btrainno_txt== train_code:   
            sleep_list.append(arvlmsg3_txt)
            train_code_list.append(btrainno_txt)
            print(f'{arvlmsg2_txt}입니다. 내립시다')
            break
               
        elif arvlcd_txt== 1  and updnline_txt== updown_condition and btrainno_txt== train_code:   
            sleep_list.append(arvlmsg3_txt)
            train_code_list.append(btrainno_txt)
            print(f'{arvlmsg2_txt}입니다. 내립시다')
            break

        elif arvlcd_txt== 2 and updnline_txt== updown_condition and btrainno_txt== train_code:
            print(f'{arvlmsg2_txt}입니다. 근데 이미 지나감 수고 이제 어플 종료')
            break
    if sleep_list==[]:
        print(return_time)
        time.sleep(30)
    else:
        #print(sleep_list)
        break
        
#  출발역 기준으로 실시간api에서 전역, 당역의 진입 혹은 도착 시 알리고 전.당역 전철 리스트로 출력
#  만약 전.당역의 진입,도착이 없을 시 다시 time.sleep()초만큼 쉬었다가 다시 확인
```
